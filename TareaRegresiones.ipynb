```python
import csv #importa libreria para leer cvs
from pandas import DataFrame #Libreria para hacer un Dataframe
import pandas as pd #Crea objeto pd
import numpy as np #Crea objeto np
import matplotlib.pyplot as plt #importa libreria para graficas 
from sklearn.linear_model import LinearRegression #Importa libreria para la regresion lineal
df1 = pd.read_csv("regresiones.csv",delimiter=';')  #lectura del archivo cvs
df=df1.drop(8) #Borrar el ultimo registro
print(df1) #imprime el arreglo
```

```python
per=df['periodos'].values.reshape(-1,1) #convierte en un arreglo los periodos 
vent1=df['ventas'].values.reshape(-1,1)#convierte en un arreglo las ventas del producto
linear_regressor=LinearRegression() #crea un objeto para la regresion linear
linear_regressor.fit(per,vent1) #hace la regresion linear de acuerdo a los periodos como x, y la Y como la salidad de cantidad
ventape=linear_regressor.predict(per) #crea un arreglo con las predicciones o las Y
m=linear_regressor.coef_[0][0] #crea coeficiente 
c= linear_regressor.intercept_[0] #crea coeficiente 
#y=mx+c 
label= r'Ventas = %0.4fper %+0.4f'%(m,c) #crea una cadena para identificar la formula
print('La ecuacion es  ',label) #imprime la formula 
cantPre=m*202209+c #Calcular la prediccion para el periodo 9
print('El valor calculado para el periodo 202209 es ',cantPre)

Error = abs((df1.iloc[8][2]-cantPre)/df1.iloc[8][2]) #calculo del error o diferencia entre valor real y el calculado
ErrorP= Error*100
print('El rango de error es de un Porcentaje de ',ErrorP)
```

```python
plt.title('Grafica para el articulo 393')
plt.scatter(202209,cantPre,color='black',label='ValorPredecido Periodo 202209')#Grafica un punto de la prediccion del periodo 9
plt.plot(df1['periodos'],df1['ventas'],color='green',label='Datos iniciales') #Grafica el comportamiento de los datos iniciales
plt.plot(per,ventape,color='yellow',label=label) #Grafica la linea de regresion lineal
plt.xlabel('periodos') #label x
plt.ylabel('cantidad') #label y
plt.legend() #Indica la descripcion de acuerdo a los colores de cada grafica
```
